```python
import pandas as pd
import numpy as np

# Set seed for reproducibility
np.random.seed(42)

# Define the number of sales representatives
num_reps = 1000

# Generate fictional dataset
data = {
    "SalesRepID": range(1, num_reps + 1),
    "SalesRepName": [f"Rep_{i}" for i in range(1, num_reps + 1)],
    "$$$ Generated": np.round(np.random.uniform(20000, 150000, num_reps), 2),
    "Leads Closed": np.random.randint(10, 500, num_reps),
    "Touches": np.random.randint(50, 2000, num_reps),
    "Conversion %": np.round(np.random.uniform(1, 50, num_reps), 2),
    "Client Meetings": np.random.randint(5, 100, num_reps),
    "Email Responses": np.random.randint(10, 300, num_reps),
    "Calls Made": np.random.randint(20, 800, num_reps),
    "Follow-ups": np.random.randint(5, 200, num_reps),
}

# Create DataFrame
df_sales_reps = pd.DataFrame(data)
df_sales_reps.head()
```

```python
# Save the DataFrame to a CSV file
df_sales_reps.to_csv('sales_representatives_dataset.csv', index=False)
```

```python
# Set seed for reproducibility
np.random.seed(42)

# Define the number of customer leads
num_leads = 1000

# Generate fictional dataset for customer leads
lead_data = {
    "LeadID": range(1, num_leads + 1),
    "LeadName": [f"Lead_{i}" for i in range(1, num_leads + 1)],
    "Contacted": np.random.choice([True, False], num_leads),
    "$$$ Potential": np.round(np.random.uniform(500, 20000, num_leads), 2),
    "Lead Source": np.random.choice(["Website", "Referral", "Cold Call", "Email Campaign", "Social Media"], num_leads),
    "Industry": np.random.choice(["Technology", "Finance", "Healthcare", "Retail", "Education"], num_leads),
    "Initial Contact Date": pd.date_range(start="2023-01-01", end="2023-12-31", periods=num_leads),
    "Last Follow-up Date": pd.date_range(start="2024-01-01", end="2024-06-25", periods=num_leads),
    "Lead Score": np.random.randint(1, 100, num_leads),
    "Interested Products": np.random.choice(["Product A", "Product B", "Product C", "Product D"], num_leads)
}

# Create DataFrame
df_customer_leads = pd.DataFrame(lead_data)

df_customer_leads.head()

df_customer_leads.to_csv('customer_leads_dataset.csv', index=False)
```
